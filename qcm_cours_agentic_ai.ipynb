```python
import random
import textwrap

LETTRES = "ABCD"

def _texte_question(q, numero=None):
    titre = f"Question {numero}" if numero is not None else "Question"
    lignes = [f"{titre} - {q['session']}", textwrap.fill(q["question"], width=88)]
    for i, choix in enumerate(q["choices"]):
        lignes.append(f"  {LETTRES[i]}. {choix}")
    return "\n".join(lignes)


def entrainement(nb_questions=10, melanger=True, correction_immediate=True):
    """Lance un QCM interactif dans Jupyter."""
    items = list(enumerate(QUESTIONS, start=1))
    if melanger:
        random.shuffle(items)
    items = items[:min(nb_questions, len(items))]

    score = 0
    total = 0
    erreurs = []

    print(f"Mode entrainement : {len(items)} questions")
    print("Tape A, B, C ou D. Tape Entree pour passer. Tape Q pour arreter.\n")

    for rang, (numero, q) in enumerate(items, start=1):
        print("=" * 80)
        print(_texte_question(q, numero=numero))

        while True:
            reponse = input("Ta reponse : ").strip().upper()
            if reponse in ("", "A", "B", "C", "D", "Q"):
                break
            print("Reponse invalide. Tape A, B, C, D, Entree ou Q.")

        if reponse == "Q":
            print("Arret de l'entrainement.")
            break

        total += 1
        bonne = LETTRES[q["answer"]]
        if reponse == bonne:
            score += 1
            print("Correct.")
        else:
            if reponse == "":
                print(f"Passe. Bonne reponse : {bonne}.")
            else:
                print(f"Faux. Ta reponse : {reponse}. Bonne reponse : {bonne}.")
            erreurs.append((numero, q, reponse or "-", bonne))

        if correction_immediate:
            print("Explication :", q["explanation"])
        print()

    print("=" * 80)
    if total == 0:
        print("Aucune question corrigee.")
        return {"score": 0, "total": 0, "erreurs": []}

    print(f"Score : {score}/{total} ({score / total:.0%})")
    if erreurs:
        print("\nA revoir :")
        for numero, q, reponse, bonne in erreurs:
            print(f"- Q{numero}: ta reponse {reponse}, bonne reponse {bonne}. {q['explanation']}")
    else:
        print("Tout est correct sur cette serie.")

    return {"score": score, "total": total, "erreurs": erreurs}


def entrainement_complet():
    """Lance toutes les questions."""
    return entrainement(nb_questions=len(QUESTIONS), melanger=True)
```
